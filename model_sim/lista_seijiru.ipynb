```python
import numpy as np
import matplotlib.pyplot as plt

n = 10
carasConsecutivas = []
for j in range(n): #vetor de 1 a n
  carasConsecutivas.append(j+1)
experimentos = 10000
tempoMedio = []

for j in range(n): #vetor de 0 a n
  tempoDeLancamento = []
  for i in range(experimentos):
    numeroDeLancamentos = 0
    numeroDeCaras = 0
    while numeroDeCaras < j+1: #continua rodando ate o numero de caras consecutivas ser igual ao j+1
      if np.random.choice([0,1]) == 1: #cara
        numeroDeCaras = numeroDeCaras + 1
      else: #coroa
        numeroDeCaras = 0
      numeroDeLancamentos += 1 #quantidade de lancamentos ate atingir o numero de caras consecutivas
    tempoDeLancamento.append(numeroDeLancamentos) #guarda o valor de cada experimento
  tempoMedio.append(sum(tempoDeLancamento)/experimentos) #guarda a media dos valores de cada experimento no vetor j
print(tempoMedio,carasConsecutivas)
plt.plot(tempoMedio,carasConsecutivas)
```

```python
import numpy as np
import matplotlib.pyplot as plt

dado = 20
experimentos = 10000
semVantagem = []
comVantagem = []

for i in range(experimentos):
  semVantagem.append(np.random.randint(dado))
  comVantagem.append(np.random.randint(dado))
  aux = np.random.randint(dado)
  if aux > comVantagem[-1]:
    comVantagem[-1] = aux

print("Dado Normal = " + str(sum(semVantagem)/experimentos))
print("Dado com Vantagem = " + str(sum(comVantagem)/experimentos))
```

```python
import numpy as np
import matplotlib.pyplot as plt

hitChanceNormal = []
hitChanceBuffed = []
dice = 20
defense = np.arange(1, 20+1)
experimentos = 10000
for i in range(int(len(defense))):
  hitChanceNormalAux = []
  hitChanceBuffedAux = []
  for j in range(experimentos):
    if np.random.randint(dice) >= i+1:
      hitChanceNormalAux.append(1)
    else:
      hitChanceNormalAux.append(0)
    if np.random.randint(dice) >= i+1:
      hitChanceBuffedAux.append(1)
    else:
      hitChanceBuffedAux.append(0)
      if np.random.randint(dice) >= i+1 :
        hitChanceBuffedAux[-1] = 1
  hitChanceNormal.append(sum(hitChanceNormalAux)/experimentos)
  hitChanceBuffed.append(sum(hitChanceBuffedAux)/experimentos)
plt.plot(defense, hitChanceNormal, color = 'r', label = 'Normal') #verde = normal
plt.plot(defense, hitChanceBuffed, color = 'g', label = 'Buffed') #vermelho = buffado
print(hitChanceNormal)
print(hitChanceBuffed)
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10
experimentos = 10000
nRange = np.arange(1, n+1)
tempoMedio = []

for i in range(n):
  tempoPorN = []
  for j in range(experimentos):
    carasTotal = 0
    tempoPorExp = 0
    while carasTotal < i+1 :
      for k in range((i+1) - carasTotal):
        carasTotal += np.random.choice([0,1])
      tempoPorExp += 1
    tempoPorN.append(tempoPorExp)
  tempoMedio.append(sum(tempoPorN)/experimentos)

plt.plot(nRange, tempoMedio)
print(tempoMedio)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

experimentos = 10000

#6-a)
contagens = []
for i in range(experimentos):
  posicao = [0,0] #[x,y]
  contagem = 0
  while posicao[0] < 5 or posicao[1] < 5:
    if np.random.choice(['x', 'y']) == 'x':
      posicao[0] += np.random.choice([-1,1])
    else:
      posicao[1] += np.random.choice([-1,1])
    contagem += 1
  contagens.append(contagem)
plt.hist(contagens)

#6-b)
n = 5 #numero de passos
resultados = []
xaxis = np.arange(1,n+1)
for i in range(n):
  distancias = []
  for j in range(experimentos):
    distancia = 0
    posicao = [0,0]
    for k in range(i+1):
      if np.random.choice(['x', 'y']) == 'x':
        posicao[0] += np.random.choice([-1,1])
      else:
        posicao[1] += np.random.choice([-1,1])
    distancia = math.sqrt(posicao[0]**2 + posicao[1]**2)
    distancias.append(distancia)
  resultados.append(sum(distancias)/experimentos)
plt.plot(xaxis, resultados)
```

```python
import numpy as np

experimentos = 100000
propriedades = [1, 3, 4, 6, 8, 9, 11, 13, 14, 17, 19, 21, 23, 26, 28, 29, 31, 33, 34, 36, 38, 39] #len == 22
rosa = [1, 3, 4]
azulclaro = [6, 8, 9]
roxo = [11, 13, 14]
laranja = [17, 19]
vermelho = [21, 23]
amarelo = [26, 28, 29]
verde = [31, 33, 34, 36]
azulescuro = [38, 39]
passagens = [0, 0, 0, 0, 0, 0, 0, 0] #len == 22
estado_atual = 0

#8-c)
for i in range(experimentos):
  estado_atual += np.random.randint(1,7) + np.random.randint(1,7)
  if estado_atual > 39:
    estado_atual -= 40 #volta completa
  if estado_atual == 30:
    estado_atual = 10 # prisao
  if estado_atual in propriedades: #verifica se o estado atual eh propriedade
    if estado_atual in rosa:
      passagens[0] += 1
    elif estado_atual in azulclaro:
      passagens[1] += 1
    elif estado_atual in roxo:
      passagens[2] += 1
    elif estado_atual in laranja:
      passagens[3] += 1
    elif estado_atual in vermelho:
      passagens[4] += 1
    elif estado_atual in amarelo:
      passagens[5] += 1
    elif estado_atual in verde:
      passagens[6] += 1
    elif estado_atual in azulescuro:
      passagens[7] += 1
print(passagens) #verde eh o melhor
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

experimentos = 1000000
lambVermelhoFC = 2
lambAzulFC = 3
def poisson(k, lamb):
  return ((math.e).math.exp(-lamb) * lamb,math.exp(k))/math.fatorial(k)
vitorias = 0
for i in range(experimentos):
  if (np.random.poisson(lambVermelhoFC) > np.random.poisson(lambAzulFC)):
    vitorias += 1
print(vitorias / experimentos)
```
